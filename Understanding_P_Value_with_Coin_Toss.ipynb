# What is a P-value and how is that used for hypothesis testing!


Let's use coin toss to gain intuition behind the this mysterious p-valu by setting up and experiment!

Suppose that you are trying to decide whether a coin is fair or biased in favor of heads. 

> The null hypothesis is 

### H0: the coin is fair (i.e., the probability of a head is 0.5), and the 

> The alternative hypothesis is 

### Ha: the coin is biased in favor of a head (i.e. the probability of a head is greater than 0.5). 


To make this problem easier, assume that the alternative hypothesis is Ha: the probability of a head is 0.7. 

You are allowed to toss the coin only 10 times, and on the basis of the outcomes, make your decision.

## P-value context


In the p-value approach neither a significance level nor a critical value are determined before the experiment is carried out or the sample taken.  The null and alternative hypotheses are stated, and the experiment is run.  A statistic is computed from the outcome of the experiment--the p-value is the probability of the observed outcome or something more extreme than the observed outcome, computed under the assumption that the null hypothesis is true.  The determination of an outcome being more extreme than the observed outcome is based on the null and alternative hypotheses.  Examples of this will be shown below.

## Simulate coin flips

This is our experiment trials

```python
import numpy as np



def coinFlip(p):    
    #perform the binomial distribution (returns 0 or 1)    
    result = np.random.binomial(1,p)
#return flip to be added to numpy array    
    return result




'''Main Area'''
#probability of heads vs. tails. This can be changed.
probability = .7
#num of flips required. This can be changed.
n = 100
#initiate array
fullResults = np.arange(n)
#perform desired numbered of flips at required probability set above
for i in range(0, n):    
    fullResults[i] = coinFlip(probability)    
    i+=1
#print results
print("probability is set to ", probability)
print("Tails = 0, Heads = 1: ", fullResults)
#Total up heads and tails for easy user experience 
print("Head Count: ", np.count_nonzero(fullResults == 1))
print("Tail Count: ", np.count_nonzero(fullResults == 0))
```

# The P-Value Approach to Hypothesis Testing

> ### We will now calculate the probabilities of coin toss under the null hypothesis that the coin indeed as fair coin.


a = the required number of successes. 

In this problem, a = 5. 

> We expect 5 heads out of 10 flips.

b = the number of trials. 

> In this problem, b = 10 = number of coin flips.

c = Number of trials-number of successes. 

In this problem, 10–5 = 5. 

> This variable represents (n-k) in the formulas above.

combinations = The number of ways to choose the success’s from the trials or the entire second part of the formula above.

percent_success = probability of flipping a head = 0.50

percent_failure = probability of no flipping a head (tail) = 0.50

```python
import math 

def binomial(trials, success):
    required = total_percent = 0
    while required <= trials:
        a = math.factorial(required)
        b = math.factorial(trials)
        c = math.factorial(trials - required)
        combinations = b / (a * c)
        percent_success = success ** required
        percent_failure = (1 - success) ** (trials - required)
        answer = combinations * percent_success * percent_failure
        total_percent += answer
        print("{} of {} instances with a {} percent likelihood of occurring.".format(required, trials, round(answer,4)))
        required += 1
    print("The total percentage ={}".format(round(total_percent,4)))
        
binomial(10, .50)
```





Ok, we completed our experiment. Recall we just concluded an experiment where the null hypothesis is that the coin is fair (p=0.5) and the alternative hypothesis is that the coin is biased in favor of heads (p>0.5). We observed through our simulation that a coin was tossed 10 times and 9 heads are observed.  Since the alternative hypothesis is p>0.5, more extreme values are numbers of heads closer to 10. 

So, to compute the p-value in this situation, we only compute the probability of 8 or more heads in 10 tosses assuming the coin is fair.  But, the number of heads in 10 tosses of a coin assuming that the coin is fair has a binomial distribution with n=10 and p=0.5. 

> ### The p-value is  P(8 heads) + P(9 heads) + P(10 heads). 
 
From the binomial probability distribution which was calculated by our fuction above, 

P(8 heads)=0.044, P(9 heads)=0.0098, and P(10 heads)=0.001.  

Therefore  the p-value is 0.0439 + 0.0098 + 0.001=0.055.

## Conclusion of our experiments



> ### Now that the p-value is computed, how do you decide whether to accept or reject the null hypothesis? 


Since the p-value is simply the probability of getting the observed number of heads under the assumption that the null hypothesis is true, ***if this probability is small, it is unlikely that the null hypothesis is true.***  

The ***'small'*** p-values lead to rejection of the null hypothesis. 

But 'small' is not defined here. We actually didn't choose the cut-off, aka alpha. In clinical trials, this alpha is used to minimize the risk of a Type I error (rejecting the null hypotheis when it is in fact true).

In practice, the definition of small is up to the experimenter--if in the opinion of the data scientist, the p-value is indeed small, the null hypothesis is rejected, while larger values would cause the null hypothesis to be accepted.  

In practice, 'small' values are usually 0.10 (90% confidence), 0.05 (95% confidence), or 0.01(99% confidence).  

In our coin tosses above, the p-value is about 0.055, and let's say a 'small' p-value for us is 0.05, ***we would fail to reject the null hypothesis***.


That is, 

> ## we would say 8 heads in 10 tosses is not enough evidence to conclude that the coin is not fair.
